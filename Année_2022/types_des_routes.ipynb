#### filter data

```
data=data.filter(col("key").isin(["highway","maxspeed","tunnel","ref","oneway","lanes","name","surface","lit","sidewalk",
                                 "bridge","bridge:name","railway","busway","cycleway"]))
```

```
data_pivot=(data.groupby("tronçon_physique2", "key", "value").count()
                       .groupby("tronçon_physique2")
                       .pivot("key")
                       .agg(first("value"))
                       
                     
         )
    
data_pivot.display()
```

```
data_final=data_pivot.withColumn("bridge", when(col("bridge").isin(["covered","viaduct","yes"]),True).otherwise(False))
data_final=data_final.withColumn("lit", when(col("lit").isin(["automatic","24/7","yes"]),True).otherwise(False))
data_final=data_final.withColumn("lanes2", when(col("lanes")>4,4).otherwise(data_final.lanes))
data_final=data_final.withColumn("oneway", when(col("oneway").isin(["alternating","-1","yes"]),"yes").otherwise("no"))
data_final=data_final.withColumn("sidewalk",when(col("sidewalk").isin(["left","both","separate","yes","right"]),"yes").otherwise("no"))
data_final=data_final.withColumn("tunnel", when(col("tunnel").isin(["building_passage","yes"]),"yes").otherwise("no"))

```

```
#surface en 3 catégories
#asphalt
#pavés:paved,sett,paving_stones,cobblestone,unhewn_cobblestone
#autres:
```

```
data_final=(data_final.withColumn("surface2", when(col("surface").isin(["paved","sett","paving_stones","cobblestone","unhewn_cobblestone"]),"pavés")
                                  .when(col("surface").isin(["asphalt"]),"asphalt")
                                  
                                     .when(col("surface").isin(["unpaved","concrete","compacted","fine_gravel","gravel","dirt"]),"autre")))
```

```
#cycleway
#opposite_lane:lane,opposite,opposite_lane
#shared:shared,shared_busway,shared_lane
#seperate: seperate,track,shoulder
```

```
data_final=(data_final.withColumn("cycleway2", when(col("cycleway").isin(["lane","opposite","opposite_lane"]),"opposite_lane")
                                  .when(col("cycleway").isin(["shared","shared_busway","shared_lane"]),"shared_lane")
                                  
                                     .when(col("cycleway").isin(["seperate","track","shoulder"]),"seperate")))
```
