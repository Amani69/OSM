```
from pyspark.sql.functions import *
import pyspark.sql.types as T
from pyspark.sql import Window
```

```
df_2022 = spark.read.parquet('/mnt/datalake/tmp/amani/data_node_way/way_2022')
df_2022=df_2022.orderBy("dateentry")
```

```
#on a 3092 nodes n'appartiennent à aucun chemin à cause de modification pour L année 2022
#dropper null
#df apres dropna  9541318
```

```
df_2022.display()
```

```
df_number_way=df_2022.groupby("nodeId").agg(count_distinct("id_way").alias("nombre_way"))
```

```
df_number_way=df_number_way.withColumn("node_intersection",when(df_number_way.nombre_way==1,False)
                                                           .otherwise(True))
```

```
df_number_way.groupby(col("node_intersection")).count().display()
```

```
data = (df_2022.join(df_number_way.select("nodeId","node_intersection"), on= df_2022.nodeId == df_number_way.nodeId,how = "left")
                            .drop(df_number_way.nodeId))
```

```
data.count()
```

```
data.display()
```

```
# si la node est la node d'intersection : je remplace le node par le node_end d'avant et way par way_id avant
windows = Window.partitionBy("rideid").orderBy(col("dateentry").asc())
df_lag = (
           data.withColumn('lag_wayid', lag('id_way').over(windows)).dropna()
              # .withColumn('lag_node_end', lag('node_end').over(windows)).dropna()
          
         )
                 
```

```
#df_lag=df_lag.withColumn("nodeId2",when(df_lag.node_intersection==True,df_lag.lag_node_end).otherwise(df_lag.nodeId))
df_lag=df_lag.withColumn("id_way2",when(df_lag.node_intersection==True,df_lag.lag_wayid).otherwise(df_lag.id_way))
```

```
df_count=df_lag.groupby(col("nodeId")).agg(count_distinct("id_way2"))
```
