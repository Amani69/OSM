```
#rcuperer la node ID  partir la colonne allnode(la deuxime node pour  la liste des nodes )
df_2022 = df_2022.withColumn('nodeId',node_selection(col("allnodes")))
```

```
def get_ways(year,spark):
  df_way = spark.read.parquet("/mnt/externaldata/france-"+year[2:]+"0101.osm.pbf.way.parquet")
  return df_way
```

```
df_ways=get_ways("2022",spark)
```

```
df_ways.display()
```

```
df_with_way = df_ways.select("id",explode("nodes").alias("nodes"))
```

```
df_with_nodes = df_with_way.select(col("id").alias("id_way"),col("nodes")["nodeId"].alias("nodeId"),col('nodes')["index"].alias("index"))
```

```
df_with_nodes.display()
```

```
data_node_way = (df_2022.join(df_with_nodes.select("nodeId","id_way"), on= df_2022.nodeId == df_with_nodes.nodeId,how = "left")
                            .drop(df_with_nodes.nodeId))
```

```
data_node_way.display()
```

```
df3=data_node_way.groupby("nodeId").agg(count_distinct("id_way"))
df3.display()
```
