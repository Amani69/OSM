```
from pyspark.sql.functions import *
import pyspark.sql.types as T
from pyspark.sql import Window
```

### Partie I recuperer toutes les donn√©es

```
all_data = spark.read.parquet("/mnt/processed-prod/Daily/SpeedRoadType/*/*/*")
all_data=all_data.orderBy("dateentry")
```

```
all_data.display()
```

```
#Convert dateentry to date
#add year 
def get_date(data):
  data=(data.withColumn("date", to_date(from_unixtime(col('dateentry')/1000)))
                     .withColumn("year",year("date"))
                     )
  return data
data_date=get_date(all_data)
```

```

#data_date.groupby("year").count().display()
```

```
#enlever les  donn√©es de 2023 
data_date=data_date.filter(col("year")!=2023)
```

```
#filter highway (les routes et les routes de liasons)
data_date=data_date.filter(col("highway").isin(["motorway","trunk","primary","secondary","tertiary","residential","unclassified","motorway_link","trunk_link","primary_link","secondary_link","tertiary_link"]))
```

```
def node_selection_f(node_list):
  """This function going to select the second node from all node (The most common one)"""  
  
  if (len(node_list) < 2):
    x= node_list[0]
  else:
    x= node_list[-1]
  return x


#The UDF to apply it on a pyspark dataframe
node_selection = udf(node_selection_f, T.LongType())
```

```
data_date= data_date.withColumn('nodeId',node_selection(col("allnodes")))
```

#### Donn√©es OSM

```
#la fonction pour recuperer les node id et way id
def get_ways(year,spark):
  df_way = spark.read.parquet("/mnt/externaldata/france-"+year[2:]+"0101.osm.pbf.way.parquet")
  return df_way

#recuperer les node id et way id pour l'ann√©e de 2022 (la derni√©re mise √† jour de donn√©es )

df_way=get_ways("2022",spark)

df_with_way = df_way.select("id",explode("nodes").alias("nodes"))
df_with_nodes = df_with_way.select(col("id").alias("id_way"),col("nodes")["nodeId"].alias("nodeId"),col('nodes')["index"].alias("index"))
```

#### jointure les donn√©es avec les donn√©es de OSM

```
#jointure pour ajouter la colonne id way dans nos donn√©es 
data_node_way = (data_date.join(df_with_nodes.select("id_way","nodeId"), on= data_date.nodeId == df_with_nodes.nodeId,how = "left").drop(df_with_nodes.nodeId))
```

#### Partie 2

```
data=spark.read.parquet('/mnt/datalake/tmp/amani/data_node_way/all_data')
data=data.orderBy("dateentry").drop("year")
```

##### Duplicate

```
#si un trajet passe par un node d'intersection  on recupere le chemin celui d'avant 
#si un trejet qui commence par un node d'intersection on recupere celui d'apr√©s
```

```
#ajouter la colonne de duplicate 
duplicate = data.groupBy("deviceid","rideId","dateentry").count().filter(col("count")>1)
duplicate = duplicate.withColumn("duplicate",lit(True))
join_date = data.join(duplicate.select("deviceid","rideId","dateentry","duplicate"),["deviceid","rideId","dateentry"],"left")
```

```
#ajouter la colonne true_way et on garde id_way ou on n' pas de duplicate si non on va mettre de null 
join_data = join_date.withColumn("duplicate",when(col("duplicate")== True,True).otherwise(False))
join_data = join_data.withColumn("true_way",when(col("duplicate")==True,None).otherwise(col("id_way")))
```

```
#on remplace soit par le chemin d'avant ou d'apres 
# remplace le lag et lead 
# mais avec dateentry asc() ou desc() pour r√©cuperer  le dernier valeur  non null
w = Window.partitionBy("rideId").orderBy(col("dateentry").asc())
join_data_2 = join_data.withColumn("true_way_2", last('true_way', True).over(w))

w = Window.partitionBy("rideId").orderBy(col("dateentry").desc())
join_data_3 = join_data_2.withColumn("true_way_3", last('true_way_2', True).over(w))
```

```
#Drop duplicate rows
df=join_data_3.dropDuplicates(["deviceid","rideId","dateentry"])
```

##### ID chaque tron√ßon physique

```
#grouper par chemin en gardant la liste de node a fin d'ajouter ID 
df_way=df.groupby("rideId","true_way_3").agg(collect_set('nodeId').alias("list_nodes"))
```

```
#grouper sans year : 323211 
#grouper avec year: 323226
#15 chemin de differnece 
# on a pas besoin de faire groupby ann√©e parce que on a toujours de trajets uniques
```

```
df_way=df_way.withColumn("tron√ßon_physique", monotonically_increasing_id())
df_way=df_way.withColumn("tron√ßon_physique2", concat(col("true_way_3"),lit("#"),col("tron√ßon_physique"))) 
```

```
#jointure final
data_tron√ßons = df.join(df_way, ["rideId","true_way_3"],"left")
```
