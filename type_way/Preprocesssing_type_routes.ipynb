```
from pyspark.sql.functions import *
import pyspark.sql.types as T
from pyspark.sql import Window
```

```
data = spark.read.parquet('/mnt/datalake/tmp/amani/data_node_way/tronçons_physiques_all_data_type')
```

```
data.display()
```

```
data.select("key").distinct().count()
```

```
#parking:lane définissent si les voies ont ou n'ont pas de place de stationnement
data=data.filter(col("key").isin(["highway","maxspeed","tunnel","tunnel:name","ref","oneway","lanes","name","surface","lit",                                            "sidewalk","bridge","bridge:name","cycleway","parking:lane"]))
```

```
data_pivot=(data.groupby("tronçon_physique2", "key", "value").count()
                       .groupby("tronçon_physique2")
                       .pivot("key")
                       .agg(first("value"))
                       
                     
         )
    
```

#### Preproccessing les valeurs de chaque caractéristique:

lanes

```
#lanes
#>4 =>4
#0=> 1
#forward=2 par 2
# pour le null : pas d'indication sur le site OSM
data_final=data_pivot.withColumn("lanes2",  when(col("lanes")>4,4)
                                           .when(col("lanes")==0,1)
                                           .when(col("lanes")=="forward=2",2)
                                           .otherwise(data_pivot.lanes))
    
```

```
data_final.groupby(col("lanes2")).count().display()
```

cycleway

```
#cycleway

data_final=(data_final.withColumn("cycleway2",  
                                  when(col("cycleway").isin(["lane","left:lane","right_lane","opposite","opposite_lane","yes",                                       "soft_lane","opposite_share_busway", "opposite_track"," designated"]),"opposite_lane")                                    .when(col("cycleway").isin(["shared","shared_busway","shared_lane","use_sidepath","sidewalk"]),"shared_lane")
                                  
                                     .when(col("cycleway").isin(["seperate","track","shoulder","left=separate","right=separate","right","left","both","cyclestreet"]),"seperate")
.when(col("cycleway").isin(["no","none","aucun","chaussidoux","asl"]),"no")))
```

```
data_final.groupby(col("cycleway2")).count().display()
```

bridge

```
data_final.groupby(col("bridge")).count().display()
```

```
data_final=data_final.withColumn("bridge", when(col("bridge").isin(["covered","viaduct","yes","movable","aqueduct","dam","cantilever"]),True).otherwise(False))
```

```
data_final.groupby(col("bridge")).count().display()
```

lit

```
data_final=data_final.withColumn("lit2", when(col("lit").isin(["automatic","24/7","yes","05:30-23:30","sunset-21:00","yes;no","P157-0633 on 2011-08-02"]),"yes")
                                        .when(col("lit").isin(["disused","no"]),"no"))
```

```
data_final.groupby(col("lit2")).count().display()
```

oneway

```
data_final=data_final.withColumn("oneway2", when(col("oneway").isin(["alternating","-1","yes"]),"yes")
                                 .when(col("oneway").isin(["service:no","no"]),"no"))
```

```
data_final.groupby(col("oneway2")).count().display()
```

sidewalk

```
data_final=data_final.withColumn("sidewalk2",when(col("sidewalk").isin(["left","both","separate","yes","right","right;both",]),"yes")
                                                .when(col("sidewalk").isin(["no","none","half","no;right"]),"no"))
```

```
data_final.groupby(col("sidewalk2")).count().display()
```

tunnel

```
data_final=data_final.withColumn("tunnel2", when(col("tunnel").isin(["building_passage","yes","avalanche_protector"]),"yes").otherwise("no"))
```

```
data_final.groupby(col("tunnel2")).count().display()
```

surface

```
data_final=(data_final.withColumn("surface2", when(col("surface").isin(["paved","sett","paving_stones","cobblestone","unhewn_cobblestone","cobblestone:flattened","pebblestone","pavement","grass_paver","pierre"]),"pavés")
                                  .when(col("surface").isin(["asphalt","asphalt_no_1","enrobes","enrobé"]),"asphalt")
                                  
                                     .when(col("surface").isin(["unpaved","concrete","compacted","fine_gravel","gravel","dirt","goudron","concrete:plates","sand","grass","earth","concrete:lanes","ground","artificial_turf","wood","metal_grid","yes"]),"autre")))
```

```
data_final.groupby(col("surface2")).count().display()
```

Parcking

```
#both,parallel,right,left,diagonal,yes,marked
data_final=data_final.withColumn("parking", when(col("parking:lane").isin(["both","parallel","right","left","diagonal","yes","marked"]),"yes")
                                 .when(col("parking:lane").isin(["no_parking","none","no"]),"no"))
```

```
data_final.groupby(col("parking")).count().display()
```
